アンケートの『自由記述』を読む

```python
# 1. 必要なライブラリをインストール（これがないと動きません）
!pip install janome wordcloud japanize-matplotlib

# 2. ライブラリの読み込み
from wordcloud import WordCloud
from janome.tokenizer import Tokenizer
import matplotlib.pyplot as plt
import japanize_matplotlib # 日本語フォントを簡単に使えるライブラリ

# 3. 分析したいテキスト（ここはアンケートのCSVなどから読み込む想定）
# 今回は例として、学校の授業アンケートの自由記述をイメージ
text = """
授業がとても楽しかったです。特に理科の実験が面白かった。
先生の話が分かりやすいです。数学が少し難しかったけれど、頑張りたい。
給食の時間が短いです。もっとゆっくり食べたい。
休み時間に図書室に行きたいけれど、本が少ないです。
体育のサッカーが楽しかった。チームで協力できたのが良かった。
宿題が多すぎて家で遊ぶ時間がありません。もう少し減らしてほしい。
"""

# 4. Janomeを使って「名詞」と「形容詞」だけを取り出す（フィルタリング）
t = Tokenizer()
words = []

for token in t.tokenize(text):
    # 品詞を取り出す
    part_of_speech = token.part_of_speech.split(',')[0]

    # 「名詞」と「形容詞」だけ抽出（助詞や「です・ます」を除外する重要な工程！）
    if part_of_speech in ['名詞', '形容詞']:
        # 不要な言葉（ストップワード）を手動で除外
        if token.surface not in ['こと', 'の', 'よう', 'それ', '私']:
            words.append(token.surface)

# スペースでつなげる（WordCloudの形式に合わせる）
text_for_wc = " ".join(words)

# 5. ワードクラウドを作成
# japanize_matplotlibを使って日本語フォントのパスを自動取得
font_path = japanize_matplotlib.get_font_ttf_path()

wordcloud = WordCloud(
    background_color="white", # 背景色
    font_path=font_path,      # 日本語フォント指定
    width=800,
    height=600,
    regexp=r"\w+"             # 1文字の単語も表示する設定
).generate(text_for_wc)

# 6. 画像として表示
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off") # 軸（枠線）を消す
plt.show()

# 画像を保存したい場合
# wordcloud.to_file("school_survey_result.png")
```
