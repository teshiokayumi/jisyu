### 【初級編-1】テスト結果の集計
ここでは、Google Colabを使って、CSV形式のテスト結果から「合計・平均・順位・偏差値」を自動計算する方法を解説します。

1. **データのアップロード**
画面左側の「フォルダアイコン」をクリックし、先ほど作成した seiseki_data.csv をドラッグ＆ドロップします。

2. **プログラムの実行**
以下のコードをコピーして、Colabのコード入力欄（セル）に貼り付け、左側の再生ボタン（▶）を押してください。 これだけで、集計からファイル出力までを一括で行います。

```python
import pandas as pd

# --- 設定：読み込むファイル名 ---
input_file = 'seiseki_data.csv'
output_file = 'seiseki_kekka.csv'

# 1. CSVファイルを読み込む
# ※Excelで作ったCSVで文字化けする場合は encoding='shift_jis' に変更してください
df = pd.read_csv(input_file, encoding='utf-8')

# 2. 点数データのみを抽出（国語から英語の列）
score_cols = ['国語', '算数', '理科', '社会', '英語']
score_df = df[score_cols]

# --- 統計量の確認（画面に表示） ---
print("【クラス全体の統計】")
print(score_df.describe())
print("-" * 30)

# 3. 合計と平均の計算
df['合計'] = score_df.sum(axis=1)
df['平均'] = score_df.mean(axis=1).round(1) # 小数点第1位で四捨五入

# 4. 順位付け（合計点の高い順）
df['順位'] = df['合計'].rank(method='min', ascending=False).astype(int)

# 5. 偏差値の計算
def calculate_deviation(series):
    mean = series.mean()
    std = series.std()
    return ((series - mean) / std * 10 + 50).round(1)

df['偏差値'] = calculate_deviation(df['平均'])

# 6. 結果を見やすく並べ替え（順位順）
df_sorted = df.sort_values('順位')

# 結果を画面に少しだけ表示
print("【集計結果（上位5名）】")
print(df_sorted.head())

# 7. CSVファイルとして保存（Excelで文字化けしない設定付き）
df_sorted.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"\n完了！ '{output_file}' という名前で保存されました。")
print("左側のフォルダアイコンからダウンロードしてください。")
```
